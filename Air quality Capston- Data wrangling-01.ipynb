**2.3.Change data type of asthma_age5_17 dataset**

 Column "Number" must change to numeric data type.But before that we need to remove(,) and (*) form the data.  Instead of using :
.str.replace('*','',regex=False)
we have to use .str.replace(r'[^0-9]', '', regex=True), because this command is designed to remove only the asterisk characters (*) from the strings. If our numbers had other non-numeric characters or spaces alongside the asterisks, these characters would remain in the string after this operation.The second command removes all characters that are not digits (0-9). Therefore, anything in the string that could potentially prevent a successful conversion to a numeric type is stripped away, leaving only the numeric part of the string.

```python
asthma_age5_17['Number'] = asthma_age5_17['Number'].str.replace(r'[^0-9]', '', regex=True)
```

```python
asthma_age5_17["Number"].isna().sum()
```

**3.Data Cleaning**

**3.1. Dropping Unneccessary Rows:**

I want to choose three neighborhoods in New York City to assess air quality that can provide an overview of the whole area requires considering diversity in geography, urban density, and potential sources of pollution. Based on the Geography column in csv files that includes all neighborhoods,I want to choose New yourk, Manhatan,Staten Island.

```python
# List of names I want to keep
rows_to_keep=["New York City","Manhattan","Staten Island"]
```

```python
aa.shape
```

```python
# Creat a new Dataframe with only rows for our study with the name,"New York City","Manhattan","Staten Island"
aa=aa.loc[aa["Geography"].isin(rows_to_keep)]

aag4= aag4.loc[aag4["Geography"].isin(rows_to_keep)]

aa5_17=aa5_17.loc[aa5_17["Geography"].isin(rows_to_keep)]

pm2_5= pm2_5.loc[pm2_5["Geography"].isin(rows_to_keep)]

no= no.loc[no["Geography"].isin(rows_to_keep)]
```

```python
aa.shape
```

**3.2.Dropping unneccessary Columns**

```python
# Dropping unneccessary Columns
aa = aa.drop(['GeoRank','Age-adjusted rate per 10,000','Estimated annual rate per 10,000'], axis=1)
aag4=aag4.drop(['GeoRank','Estimated annual rate per 10,000'], axis=1)
aa5_17=aa5_17.drop(['GeoRank','Estimated annual rate per 10,000'], axis=1)
pm2_5=pm2_5.drop(['GeoRank','10th percentile mcg/m3','90th percentile mcg/m3'], axis=1)
no=no.drop(['GeoRank','10th percentile ppb','90th percentile ppb'], axis=1)
```

```python
print("aa:", aa.shape)
print("aag4:", aag4.shape)
print("aa5_17:", aa5_17.shape)
print("pm2_5:",pm2_5.shape)
print("no:",no.shape)
```

```python
print(aa.columns)
```

```python
#rename the column 'TimePeriod' to 'Year'
aa= aa.rename(columns={"TimePeriod":"Year"})

aag4=aag4.rename(columns={"TimePeriod":"Year"})

aa5_17=aa5_17.rename(columns={"TimePeriod":"Year"})

pm2_5= pm2_5.rename(columns={"TimePeriod":"Year"})

no= no.rename(columns={"TimePeriod":"Year"})
```

**4.Joining Data**

```python
merge1=pd.merge(aa,aag4, on=["Year", "GeoType", "GeoID"], how="inner")
```

```python
merge2= pd.merge(merge1,aa5_17, on=["Year", "GeoType", "GeoID"], how="inner")
```

```python
# I got error to merge on  mismatch in data types, so I have to check the data type
print(merge2[['Year', 'GeoType', 'GeoID']].dtypes)
print(pm2_5[['Year', 'GeoType', 'GeoID']].dtypes)
```

```python
# changed dtype to int
pm2_5['Year'] = pm2_5['Year'].astype(int)
```

```python
merge3= pd.merge(merge2,pm2_5, on=["Year", "GeoType", "GeoID"], how="inner",suffixes=('_merge2', '_pm25'))
```

```python
merge4=pd.merge(merge3,no,on=["Year", "GeoType", "GeoID"], how="inner")
```
