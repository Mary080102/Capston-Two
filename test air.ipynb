```python
# to merge datasets based on GeoID and TimePeriod, both of them must have a 
#same data type so I have to delete prefix Annual Average from TimePeriod column

#Remove the 'Annual' prefix from the 'TimePeriod' in no2_sub_1
no2_sub_1 = no2_sub_1.copy()
no2_sub_1["TimePeriod"] = no2_sub_1["TimePeriod"].str.replace("Annual Average", "")
```

```python
# to merge datasets based on GeoID and TimePeriod, both of them must have a 
#same data type so I have to delete prefix Annual Average from TimePeriod column

#Remove the 'Annual' prefix from the 'TimePeriod' in pm_2_5_sub_1
pm_2_5_sub_1 = pm_2_5_sub_1.copy()

pm_2_5_sub_1["TimePeriod"] = pm_2_5_sub_1["TimePeriod"].str.replace("Annual Average", "")
```

```python
pm_2_5_sub_1.head()
```

```python
# Based on the resulte from .info() columns of "Timeperiod in no2  and pm_2_5 datasets were object that must be int
#to be matched with "GeoID" column.
#now we can match the data type of both columns "TimePeriod" and "GeoID"
no2_sub_1["TimePeriod"]= no2_sub_1["TimePeriod"].astype(int)
pm_2_5_sub_1["TimePeriod"]=pm_2_5_sub_1["TimePeriod"].astype(int)
```

```python
# Drop unneccessary columns
#List the columns that want to drop
columns_to_drop=["GeoType","GeoRank"]
```

```python
## Create a new DataFrame with only the necessary columns
asthma_adults_sub_drop= asthma_adults_sub.drop(columns_to_drop, axis=1)
```

```python
asthma_age_4_sub_drop= asthma_age_4_sub.drop(columns_to_drop, axis=1)
```

```python
no2_sub_1_drop=no2_sub_1.drop(columns_to_drop, axis=1)
```

```python
pm_2_5_sub_1_drop=pm_2_5_sub_1.drop(columns_to_drop, axis=1)
```

```python
#joining data
```

```python
# rename the common columns except 'GeoID' to merge dataframes
asthma_adults_sub_drop= asthma_adults_sub_drop.rename(columns={'Geography':'Geography_ad'})

asthma_age_4_sub_drop= asthma_age_4_sub_drop.rename(columns={'Geography':'Geography_ag_4'})

asthma_age5_17_sub_drop= asthma_age5_17_sub_drop.rename(columns={'Geography':'Geography_ag5_17'})

no2_sub_1_drop=no2_sub_1_drop.rename(columns={'Geography':'Geography_ag5_17'})

pm_2_5_sub_1_drop=pm_2_5_sub_1_drop.rename(columns={'Geography':'Geography_ag5_17'})
```

```python
# merge filtered DataFrames
merge1= pd.merge(asthma_adults_sub_drop,asthma_age_4_sub_drop,on=["GeoID","TimePeriod"], how= "inner")
```

```python
merge2=pd.merge(merge1,asthma_age5_17_sub_drop, on=["GeoID","TimePeriod"], how= "inner")
```

```python
merge3= pd.merge(merge2,no2_sub_1_drop,on=["GeoID","TimePeriod"], how= "inner")
```

```python
merge4=pd.merge(merge3,pm_2_5_sub_1_drop,on=["GeoID","TimePeriod"], how= "inner")
```
