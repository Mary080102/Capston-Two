### **2.3.Change data type of asthma_age5_17 dataset**

 Column "Number" must change to numeric data type.But before that we need to remove(,) and (*) form the data. (*) in dataset means that Estimate is based on small numbers so should be interpreted with caution. So I want to add new column to asthma_age5_17 that will indicate whether the original value had an asterisk. This way, I can maintain the information about the reliability of the estimate.

```python
# Using .value_counts()to see if there are any unusual values in column"Estimated annual rate per 10,000".
asthma_age5_17["Number"].value_counts()
```

```python
asthma_age5_17['Number'].dtype
```

```python
# This will show the types of data present in the column
print(asthma_age5_17['Number'].apply(type).value_counts())
```

 Instead of using :
.str.replace('*','',regex=False)
we have to use .str.replace(r'[^0-9]', '', regex=True), because this command is designed to remove only the asterisk characters (*) from the strings. If our numbers had other non-numeric characters or spaces alongside the asterisks, these characters would remain in the string after this operation.The second command removes all characters that are not digits (0-9). Therefore, anything in the string that could potentially prevent a successful conversion to a numeric type is stripped away, leaving only the numeric part of the string.

```python
asthma_age5_17['Number'] = asthma_age5_17['Number'].str.replace(r'[^0-9]', '', regex=True)
```

```python
asthma_age5_17["Number"].isna().sum()
```

```python
#convert column to numeric, setting errors='coerce'
asthma_age5_17["Number"]= pd.to_numeric(asthma_age5_17["Number"], errors="coerce")
```

```python
asthma_age5_17['Number'].dtype
```

```python
# Check for empty strings after replacement
print("Empty strings after replacement:", (asthma_age5_17['Number'] == '').sum())
```
